```python
import random
from collections import Counter

stats = []



class Door:
    def __init__ (self):
        self.is_lambo = False
    def reveal(self):
        return self.is_lambo
    

class Door_factory:
    def __init__(self, number):
        self.number = number

    def make_doors (self):
        door_list = []
        for i in range(self.number):
            door_list.append(Door())
        winning_door = random.choice (door_list)
        winning_door.is_lambo =True
        return door_list      


class Participant:
    def __init__(self, door_list, strategy):
        self.door_list = door_list
        self.strategy = strategy

    def choose(self):
        self.chosen_door = random.choice (self.door_list)
        self.door_list.remove (self.chosen_door)
        return self.chosen_door

    def action(self):
        self.strategy.execute(self)
        
        
class Strategy:
    def __init__(self):
        pass

class StrategySwitch(Strategy):

    def execute(self, participant):
        switch = random.choice(participant.door_list)
        print(switch.is_lambo)
        stats.append(switch.is_lambo)

class StrategyKeep(Strategy):

    def execute(self, participant):
        keep = participant.chosen_door
        print(keep.is_lambo)
        stats.append(keep.is_lambo)

class StrategyRandom(Strategy):

    def execute(self, participant):
        strategy_pick = random.choice([StrategyKeep, StrategySwitch])
        strategy = strategy_pick()
        strategy.execute(participant)



class Monty:
    def __init__(self, door_list):
        self.door_list = door_list
    def reveal(self):
        revealed_door = random.choice (self.door_list)
        if revealed_door.is_lambo:
            self.reveal()
        else:
            self.door_list.remove(revealed_door)

def pre_round():
    door_factory = Door_factory(3)
    list_of_doors = door_factory.make_doors()
    strategy = StrategyKeep()
    hmon = Participant(list_of_doors, strategy)
    monty = Monty(list_of_doors)
    return hmon, monty

def round(hmon,monty):
    hmon.choose()
    monty.reveal()
    hmon.action()


for i in range(100_000):
    hmon,monty = pre_round()
    round(hmon,monty)

print(Counter (stats))
```
